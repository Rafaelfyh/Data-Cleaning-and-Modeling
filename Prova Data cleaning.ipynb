```python
df=pd.read_csv('winemag-data-130k-v2.csv')
```

```python
# descobrir o que a função faz de acordo com o codigo da tarefa 
# importar sklearn importando KBinsDiscretizer
#variavel train faz esclusao dos valores de acordo com dropna com somente valores, com reshape deixa em matriz.
#discretizer chama  a funcao kbindiscretizer
# fit treinamento ultilizo ele con discretizer.fit(train)
# criacao coluna points group e transformo coluna baseado no já feito 


from sklearn.preprocessing import KBinsDiscretizer
train = df.points.dropna().values.reshape(-1,1)
discretizer =KBinsDiscretizer(n_bins=4,encode="ordinal",strategy='uniform')

discretizer.fit(train)


df['points_group']=discretizer.transform(train)*5+80
```

```python
#retornar valores point group

df['points_group'].value_counts()
```

```python
# criar variavel df price per group depois dataframe chamando colunas
#agrupou e pegou media da price

df_price_per_group = df[["country", "points_group","price"]].groupby(["country", "points_group"]).mean() 

df_price_per_group
```

```python
#retornar price e alterar nome da coluna

df_price_per_group.columns=['price_per_group']
```

```python
# refazer o indice
# subscrever df price

df_price_per_group.reset_index()
df_price_per_group=df_price_per_group.reset_index()
```

```python
# merge da esquerda onde country e points grupos

df_merge=df.merge(df_price_per_group,how='left',on=['country','points_group'])
df_merge
```

```python
# price do merge onde for fillna price por group e tirou a media
 
df_merge['price'].fillna(df_merge['price_per_group']).mean()
```

```python
import numpy as np
# inicia uma lista vazia onde colocaremos os outliers
outliers = []
df['price']=df['price'].dropna()

# desvio padrão 
desv_pad = np.std(df['price'])

# média dos seus dados
media = np.mean(df['price'])

# coloca o limite que determina se é outlier
limite = desv_pad * 3

limite_inferior = media - limite
limite_superior = media + limite

# Localiza outliers e anexa à nossa lista que iniciamos sem nada
for outlier in df['price']:
    if outlier > limite_superior or outlier < limite_inferior:
        outliers.append(outlier)


# conta quantos elementos são outliers
print(len(outliers))

```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import re
from nltk.tokenize import word_tokenize
from nltk.tokenize import TreebankWordTokenizer, WordPunctTokenizer, WhitespaceTokenizer
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
import string
import re
snowball_stemmer = SnowballStemmer('portuguese')

def clean_text(sentence):
    word_tokens= word_tokenize(sentence)
    new_words= [word for word in word_tokens if (word not in string.punctuation and not re.match("[0-9]+", word))]
    clean_tokens = new_words[:]
    for token in clean_tokens:
        if token in stopwords.word('english'):
            clean_tokens.remove(token)
    return ' '.join(clean_tokens)  
    
clean_text(df['description'])

lista = [x for x in df['description']]
nova_lista=[]
for i in lista:
    nova_lista.append(clean_text(i))
nova_lista
   
```
